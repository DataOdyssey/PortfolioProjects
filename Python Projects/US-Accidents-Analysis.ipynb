# Analyzing Start Time column

---



```python
df.Start_Time
```

```python
df.Start_Time = pd.to_datetime(df.Start_Time)     ## Parse Dates
```

```python
df.Start_Time[0]
```

```python
sns.distplot(df.Start_Time.dt.hour, bins=24, kde = False, norm_hist = True)   ## Extracting Hour from timestamp and plotting it with seaborn with 24 bins
```

```python
sns.distplot(df.Start_Time.dt.dayofweek, bins=7, kde = False, norm_hist = True) 
```

## Is the distribution of accidents by hour the same on weekends as on weekdays?

---



```python
sundays_start_time = df.Start_Time[df.Start_Time.dt.dayofweek == 6]
sns.distplot(sundays_start_time.dt.hour, bins=24, kde = False, norm_hist = True)    ## Hourly Basis
```

```python
monday_start_time = df.Start_Time[df.Start_Time.dt.dayofweek == 0]      ## Weekly Basis
sns.distplot(monday_start_time.dt.hour, bins=24, kde = False, norm_hist = True) 
```

```python
df_2019 = df[df.Start_Time.dt.year == 2016]           ## Yearly Basis   Year = 2016
sns.distplot(df_2019.Start_Time.dt.month, bins = 12, kde = False, norm_hist = True)
```

```python
df_2019 = df[df.Start_Time.dt.year == 2017]     ## Year = 2017
sns.distplot(df_2019.Start_Time.dt.month, bins = 12, kde = False, norm_hist = True)
```

```python
df_2019 = df[df.Start_Time.dt.year == 2018]     ## Year = 2018
sns.distplot(df_2019.Start_Time.dt.month, bins = 12, kde = False, norm_hist = True)
```

```python
df_2019 = df[df.Start_Time.dt.year == 2019]     ## Year = 2019
sns.distplot(df_2019.Start_Time.dt.month, bins = 12, kde = False, norm_hist = True)
```

```python
df_2019 = df[df.Start_Time.dt.year == 2020]     ## Year = 2020
sns.distplot(df_2019.Start_Time.dt.month, bins = 12, kde = False, norm_hist = True)
```



---


## Inference :-
- A high percentage of accidents occur between 6 am to 10 am (probably people in a hurry to get to work)
- Next highest percentage is 3 pm to 6pm
- The distribution of accidents by hour is the same on weekends as on weekdays
- But, Fewer accidents on weekends (No rush hour traffic as well)
- More number of accidents in winter months
- Much of the data is missing for 2016 and 2020
